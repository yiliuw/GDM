```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
list1 = np.array([x[0] for x in slds])
list2 = np.array([x[0] for x in slds_test])
list3 = np.array(slds_score)

lyst1 = np.array([x[0] for x in gdm])
lyst2 = np.array([x[0] for x in gdm_test])
lyst3 = np.array(gdm_score)


# Build tidy dataframe
values = np.concatenate([list1, list2, list3, lyst1, lyst2, lyst3])
groups = (["Train R2"]*len(list1) + ["Test R2"]*len(list2) + ["State Quality"]*len(list3) +
          ["Train R2"]*len(lyst1) + ["Test R2"]*len(lyst2) + ["State Quality"]*len(lyst3))
sets = (["slds"]*(len(list1)+len(list2)+len(list3)) +
        ["gdm"]*(len(lyst1)+len(lyst2)+len(lyst3)))

df = pd.DataFrame({"value": values, "group": groups, "set": sets})

# Plot params
group_order = ["Train R2", "Test R2", "State Quality"]
hue_order = ["slds", "gdm"]
dot_palette = {"slds": "black", "gdm": "dodgerblue"}
box_palette = {"slds": "lightgray", "gdm": "lightgray"}

plt.figure(figsize=(7,4))


# Boxplot (no fliers, no dodge issues)
sns.boxplot(
    x="group", y="value", hue="set", data=df,
    order=group_order, hue_order=hue_order,
    width=0.8, palette=box_palette, dodge=True,
    showfliers=False, legend=False
)

# Stripplot (force align with boxplot dodge)
sns.stripplot(
    x="group", y="value", hue="set", data=df,
    order=group_order, hue_order=hue_order,
    dodge=True, jitter=0.2, alpha=0.6, size=5,
    palette=dot_palette, legend=False
)

plt.xlabel("")
plt.ylabel("")
plt.legend([], [], frameon=False)
plt.ylim(0, 1)
plt.title("")
plt.tight_layout()
plt.show()
```
